```python
Rank = list(range(1, 16601))
Global_Sales = [82.74 - (i * 0.01) for i in range(16600)]

# Create a scatter plot
plt.scatter(Rank, Global_Sales, s=5, alpha=0.5, color='b')

# Add labels and title
plt.xlabel('Rank')
plt.ylabel('Global Sales (in millions)')
plt.title('Scatterplot of Rank vs. Global Sales')

# Reverse the y-axis to match the specified range
plt.gca().invert_yaxis()

# Show the plot
plt.show()





```
